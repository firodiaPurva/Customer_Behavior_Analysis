```python
df['REFUNDED_AMOUNT'] = df['REFUNDED_AMOUNT'].str.replace('$', '')
df['SUB_TOTAL'] = df['SUB_TOTAL'].replace('[\$,]', '', regex=True).astype(float)
df['DELIVERY_FEE'] = df['DELIVERY_FEE'].replace('[\$,]', '', regex=True).astype(float)
df['SERVICE_FEE'] = df['SERVICE_FEE'].replace('[\$,]', '', regex=True).astype(float)
df['DISCOUNT'] = df['DISCOUNT'].replace('[\$,]', '', regex=True).astype(float)
df['TIP'] = df['TIP'].replace('[\$,]', '', regex=True).astype(float)
```

```python
def time_difference(row):
    order_time = row['TIME_ORDER_PLACED_AT_RESTAURANT']
    total_time = row['DELIVERY_TIME']

    time_diff_seconds = (total_time.hour * 3600 + total_time.minute * 60 + total_time.second) - \
                        (order_time.hour * 3600 + order_time.minute * 60 + order_time.second)

    return time_diff_seconds
```

```python
def calculate_time_difference(row):
    order_time = row['TIME_ORDER_PLACED_AT_RESTAURANT']
    driver_time = row['TIME_DRIVER_ARRIVED_AT_RESTAURANT']

    time_diff_seconds = (driver_time.hour * 3600 + driver_time.minute * 60 + driver_time.second) - \
                        (order_time.hour * 3600 + order_time.minute * 60 + order_time.second)

    return time_diff_seconds
```

```python
df['TIME_ORDER_PLACED_AT_RESTAURANT'] = pd.to_datetime(df['TIME_ORDER_PLACED_AT_RESTAURANT'], format='%H:%M:%S').dt.time
df['TIME_DRIVER_ARRIVED_AT_RESTAURANT'] = pd.to_datetime(df['TIME_DRIVER_ARRIVED_AT_RESTAURANT'], format='%H:%M:%S').dt.time
df['DELIVERY_TIME'] = pd.to_datetime(df['DELIVERY_TIME'], format='%H:%M:%S').dt.time

df['TOTAL_TIME_TAKEN_SECONDS'] = df.apply(time_difference, axis=1)

df['TOTAL_TIME_TAKEN_MINUTES'] = df['TOTAL_TIME_TAKEN_SECONDS'] / 60

df[['TIME_ORDER_PLACED_AT_RESTAURANT', 'TIME_DRIVER_ARRIVED_AT_RESTAURANT', 'TOTAL_TIME_TAKEN_SECONDS','TOTAL_TIME_TAKEN_MINUTES']].head()
```

```python
# Removing negative values
df = df[df['TOTAL_TIME_TAKEN_MINUTES'] >= 0]
```

```python
df.reset_index(drop=True, inplace=True)
```

### 2. Customer segments based on their ordering habits

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder, StandardScaler


label_encoder = LabelEncoder()
df['DeliveryTimeEncoded'] = label_encoder.fit_transform(df['DELIVERY_TIME'])

features = df[['DeliveryTimeEncoded']]

scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

optimal_k = 2 #Elbow method below suggests 2 is optimum number of clusters

kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df['DeliveryTimeSegment'] = kmeans.fit_predict(features_scaled)

segment_mapping = {
    0: 'ASAP Delivery',
    1: 'Scheduled Delivery'
}

df['DeliveryTimeSegment'] = df['DeliveryTimeSegment'].map(segment_mapping)

print(df[['CUSTOMER_ID', 'DELIVERY_TIME', 'DeliveryTimeSegment']])
```

```python
# PErforming Elbow method to find out optimum number of clusters
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder


k_values = range(1, 11)

inertia_values = []
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features_scaled)
    inertia_values.append(kmeans.inertia_)

# Plot the elbow curve
plt.plot(k_values, inertia_values, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

- As per the above plot, we can say the two is the optimal number of clusters.

### 3.Segments by Time-of-Day :

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

df["Date"]=pd.to_datetime(df['DATE'])


df['OrderDateTime'] = df['DATE'] + pd.to_timedelta(df['TIME_CUSTOMER_PLACED_ORDER'].astype(str))

df['OrderHour'] = df['OrderDateTime'].dt.hour

label_encoder = LabelEncoder()
df['OrderHourEncoded'] = label_encoder.fit_transform(df['OrderHour'])

features = df[['OrderHourEncoded']]

# Choosing the optimal number of clusters (e.g., 4 for breakfast, lunch, dinner, late-night)
optimal_k = 4

kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df['TimeOfDaySegment'] = kmeans.fit_predict(features)

time_of_day_mapping = {
    0: 'Breakfast',
    1: 'Lunch',
    2: 'Dinner',
    3: 'Late-Night'
}

df['TimeOfDaySegment'] = df['TimeOfDaySegment'].map(time_of_day_mapping)

print(df[['DATE', 'TIME_CUSTOMER_PLACED_ORDER', 'OrderHour', 'TimeOfDaySegment']])
```

```python
# Visualizing the segments
sns.countplot(x='TimeOfDaySegment', data=df, palette='pastel')
plt.title('Distribution of Orders Across Time-of-Day Segments')
plt.show()
```

### Churn Prediction

#### Usuing Random Forest

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import RepeatedStratifiedKFold
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler

df['Recency'] = (pd.to_datetime('2020-04-30') - pd.to_datetime(df['DATE'], format='%m/%d/%Y')).dt.days
df['OrderFrequency'] = df.groupby('CUSTOMER_ID')['DATE'].transform('count')
df['TotalSpending'] = df.groupby('CUSTOMER_ID')['SUB_TOTAL'].transform('sum')

# Eliminating the feature Total Tip
#df['TotalTip']= df.groupby('CUSTOMER_ID')['TIP'].transform('sum') 

# Creating labels for churn prediction
df['Churn'] = df.groupby('CUSTOMER_ID')['DATE'].transform('max') > pd.to_datetime('2020-04-30') - pd.DateOffset(days=45)  # Churn defined as no order in the last 30 days

# Selecting relevant features and labels
features = df[['Recency', 'OrderFrequency','TotalSpending']]
labels = df['Churn']

X = features
y = df['Churn']

# Balancing the dataset using undersampling (you can choose either undersampling or oversampling)
undersampler = RandomUnderSampler(sampling_strategy = 1)
X_balanced, y_balanced = undersampler.fit_resample(X, y)

imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_balanced)

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_train_imputed, y_balanced, test_size=0.2, random_state=42, stratify=y_balanced)

# Building a RandomForestClassifier model
model = RandomForestClassifier(n_estimators=500, random_state=42)
model.fit(X_train, y_train)

# Making predictions on the test set
y_pred = model.predict(X_test)

# Evaluating the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

# Evaluation metrics
print(f'Accuracy: {accuracy:.2f}')
print('\nConfusion Matrix:')
print(conf_matrix)
print('\nClassification Report:')
print(class_report)
```

2) Class Imbalance: 
A balancing method (in this case, undersampling) is applied. A common issue with churn prediction datasets is class imbalance, which occurs when there are significantly fewer churning customers than remaining customers.
